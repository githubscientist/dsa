```python
def findFlies(frog, tongueSize, Y):
    # initialize the number of flies that the frog can eat
    flies = 0

    # iterate the flies list Y
    for i in range(len(Y)):
        # for every fly, check whether the frog can eat it  
        if frog + tongueSize >= Y[i]:
            # if true => increase the number of flies
            flies += 1
    
    # return the number of flies
    return flies


def frogFeast(X, S, Y):
    # create an empty list to store the counts of flies <- fliesList
    fliesList = []

    # Iterate the list X <- i
    for i in range(len(X)):
        # For every frog at X[i], 
        # find the number of flies the frog can eat <- flies
        flies = findFlies(X[i], S[i], Y)

        # append the flies to the fliesList
        fliesList.append(flies)

    #print the fliesList
    print(fliesList)
```
