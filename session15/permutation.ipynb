```python
# APPROACH #1
# complete the following function that prints the minimum cost
# to reach from the given array to the permutation array
def permutationCost(numbers):
    # find the length of the given list numbers <- N
    # Test case #1 : N = 4
    N = len(numbers)

    # create a list with values from 1 to N <- permutation
    # permutation = [number for number in range(1, N+1)]

    permutation = []

    for number in range(1, N+1):
        permutation.append(number)

    diffSum = 0

    # find the sum of differences between the elements of numbers and permutation <- diffSum
    # iterate from 0 index to N-1
    for index in range(0, N):
        diffSum += (numbers[index] - permutation[index])
    
    # print the sum of differences (diffSum)
    print(diffSum)
```

```python
# APPROACH #2

def permutationCost(numbers):
    # get the length of the numbers <- N
    N = len(numbers)

    # create a permutation list with the numbers from 1 to N <- permutation
    permutation = [number for number in range(1, N+1)]

    # find the sum of numbers list
    numbersSum = 0
    for i in range(len(numbers)):
        numbersSum += numbers[i]

    # find the sum of permutation list
    permutationSum = 0
    for i in range(len(permutation)):
        permutationSum += permutation[i]

    # print the absolute difference between both the sums
    print(abs(numbersSum - permutationSum))
```
